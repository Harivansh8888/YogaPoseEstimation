***Displaying a random image from Dataset***

```python
import random
import matplotlib.pyplot as plt




def random_images():
    random_classname = random.choice(class_names)
    p = random_classname
    plt.figure(figsize=(10,10))

    path = "/content/drive/MyDrive/kaggle_dataset/dataset"
    inner_path = path+"/"+p
    files = os.listdir(inner_path)
    d = random.choice(files)
    image_path = inner_path+"/"+d
    print(image_path)

    image = plt.imread(image_path)
    plt.imshow(image)
    plt.axis(False)

    print(image.shape)
    print("Max: ", image.max())
    print("Min: ", image.min())


random_images()
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
tf.random.set_seed(42)

dir = "/content/drive/MyDrive/kaggle_dataset/dataset"

datagen = ImageDataGenerator(rescale=1/255.,
                             width_shift_range=0.2,
                             height_shift_range=0.2,
                             shear_range=0.2,
                             rotation_range=20,
                             horizontal_flip = True)

data = datagen.flow_from_directory(directory=  dir,
                                   target_size= (512, 512),
                                   class_mode= 'categorical',
                                   batch_size= 32,
                                   shuffle= True,)
```

```python
images, labels = data.next()
random_number = random.sample(range(len(images)), 1)[0]
plt.imshow(images[random_number])
#plt.title(class_names[int(labels[random_number])])
plt.axis(False);
```
