***Data Augmentation***

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
tf.random.set_seed(42)

dir = "/content/drive/MyDrive/kaggle_dataset/dataset"

datagen = ImageDataGenerator(rescale=1/255.,
                             width_shift_range=0.2,
                             height_shift_range=0.2,
                             shear_range=0.2,
                             rotation_range=20,
                             horizontal_flip = True)

data = datagen.flow_from_directory(directory=  dir,
                                   target_size= (512, 512),
                                   class_mode= 'categorical',
                                   batch_size= 32,
                                   shuffle= True,)
```

***Creating the Base Model***

```python
from tensorflow.keras import layers
from tensorflow.keras.layers.experimental import preprocessing
from tensorflow.keras.models import Sequential
```

```python
base_model = tf.keras.applications.EfficientNetB0(include_top=False)
base_model.trainable=False

# Setup model architecture with trainable top layers
inputs = layers.Input(shape=(512,512,3), name="input_layer")
x = base_model(inputs, training=False) #Put the base model in inference mode, and so weights ehich remains frozen, remains frozen
x = layers.GlobalAveragePooling2D(name="global_average_pool_layer")(x)
outputs = layers.Dense(len(class_names), activation="sigmoid", name="output_layer")(x)
model = tf.keras.Model(inputs, outputs)
```

```python
model.summary()
```

```python
base_model.summary()
```

***Compiling the Base Model***

```python
# Compile 
model.compile(loss="categorical_crossentropy",
              optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),
              metrics=["accuracy"])

# Fit
history = model.fit(data,
                    epochs=10,
                    steps_per_epoch=len(data))
```

***Creating the Model***

```python
base_model_2 =  tf.keras.applications.InceptionResNetV2(
    include_top = False, 
    weights = 'imagenet', 
    input_tensor = None,
    input_shape = (512, 512, 3), 
    pooling = None, 
    classes = 1000
)
```

```python
base_model_2.summary()
```

```python
base_model_2.trainable=False
```

```python
# Setup model architecture with trainable top layers
inputs = layers.Input(shape=(512,512,3), name="input_layer")
x = base_model_2(inputs, training=False) #Put the base model in inference mode, and so weights ehich remains frozen, remains frozen
x = layers.Conv2D(64, (3, 3), activation = 'relu')(x)
x = layers.Flatten()(x)
outputs = layers.Dense(len(class_names), activation="softmax", name="output_layer")(x)
model = tf.keras.Model(inputs, outputs)
```

***Compiling the Model***

```python
model.compile(loss = 'categorical_crossentropy', 
              optimizer = tf.keras.optimizers.Adam(learning_rate=0.0001), 
              metrics = ["accuracy"])
```

```python
# Fit
history = model.fit(data,
                    epochs=10,
                    steps_per_epoch=len(data))
```

***Functions for Predicting the Result when an image is provided***

```python
def load_and_prep_image(filename, img_shape=512, scale=True):

  img = tf.io.read_file(filename)
  img = tf.image.decode_image(img, channels=3)
  img = tf.image.resize(img, [img_shape, img_shape])
  if scale:
    return img/255.
  else:
    return img
```

```python
import matplotlib.pyplot as plt

def pred_and_plot(model, filename, class_names):

  img = load_and_prep_image(filename)
  pred = model.predict(tf.expand_dims(img, axis=0))

  if len(pred[0]) > 1: 
    pred_class = class_names[pred.argmax()]
  else:
    pred_class = class_names[int(tf.round(pred)[0][0])]

  plt.imshow(img)
  plt.title(f"Prediction: {pred_class} \nProbability: {pred.max()*100:.3f}%")
  plt.axis(False);
```

```python
import itertools
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import confusion_matrix
```

```python
def make_confusion_matrix(y_true, y_pred, classes=None, figsize=(10, 10), text_size=15, norm=False, savefig=False): 
  
  cm = confusion_matrix(y_true, y_pred)
  cm_norm = cm.astype("float") / cm.sum(axis=1)[:, np.newaxis]
  n_classes = cm.shape[0]

  
  fig, ax = plt.subplots(figsize=figsize)
  cax = ax.matshow(cm, cmap=plt.cm.Blues) 
  fig.colorbar(cax)

  labels = classes
  
  
  
  ax.set(title="Confusion Matrix",
         xlabel="Predicted label",
         ylabel="True label",
         xticks=np.arange(n_classes),
         yticks=np.arange(n_classes), 
         xticklabels=labels,
         yticklabels=labels)
  

  ax.xaxis.set_label_position("bottom")
  ax.xaxis.tick_bottom()

  plt.xticks(rotation=70, fontsize=text_size)
  plt.yticks(fontsize=text_size)



  threshold = (cm.max() + cm.min()) / 2.

  
  for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
    if norm:
      plt.text(j, i, f"{cm[i, j]} ({cm_norm[i, j]*100:.1f}%)",
              horizontalalignment="center",
              color="white" if cm[i, j] > threshold else "black",
              size=text_size)
    else:
      plt.text(j, i, f"{cm[i, j]}",
              horizontalalignment="center",
              color="white" if cm[i, j] > threshold else "black",
              size=text_size)

  
  if savefig:
    fig.savefig("confusion_matrix.png")
```

```python
make_confusion_matrix(y_true=y_labels,
                      y_pred=pred_classes,
                      classes=class_names,
                      figsize=(100,100),
                      text_size=20,
                      savefig=True)
```
