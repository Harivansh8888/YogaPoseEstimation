***Data Augmentation***

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
tf.random.set_seed(42)

dir = "/content/drive/MyDrive/kaggle_dataset/dataset"

datagen = ImageDataGenerator(rescale=1/255.,
                             width_shift_range=0.2,
                             height_shift_range=0.2,
                             shear_range=0.2,
                             rotation_range=20,
                             horizontal_flip = True)

data = datagen.flow_from_directory(directory=  dir,
                                   target_size= (512, 512),
                                   class_mode= 'categorical',
                                   batch_size= 32,
                                   shuffle= True,)
```

***Displaying a random image from the Augmented Dataset***

```python
images, labels = data.next()
random_number = random.sample(range(len(images)), 1)[0]
plt.imshow(images[random_number])
plt.title(class_names[random_number])
plt.axis(False);
```

```python
from tensorflow.keras import layers
from tensorflow.keras.layers.experimental import preprocessing
from tensorflow.keras.models import Sequential
```

```python
base_model = tf.keras.applications.EfficientNetB0(include_top=False)
base_model.trainable=False

# Setup model architecture with trainable top layers
inputs = layers.Input(shape=(512,512,3), name="input_layer")
x = base_model(inputs, training=False) #Put the base model in inference mode, and so weights ehich remains frozen, remains frozen
x = layers.GlobalAveragePooling2D(name="global_average_pool_layer")(x)
outputs = layers.Dense(len(class_names), activation="sigmoid", name="output_layer")(x)
model = tf.keras.Model(inputs, outputs)
```

```python
model.summary()
```

```python
base_model.summary()
```

```python
# Compile 
model.compile(loss="categorical_crossentropy",
              optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),
              metrics=["accuracy"])

# Fit
history = model.fit(data,
                    epochs=10,
                    steps_per_epoch=len(data))
```

```python
model.summary()
```

```python
base_model_2 =  tf.keras.applications.InceptionResNetV2(
    include_top = False, 
    weights = 'imagenet', 
    input_tensor = None,
    input_shape = (512, 512, 3), 
    pooling = None, 
    classes = 1000
)
```

```python
base_model_2.summary()
```

```python
base_model_2.trainable=False
```

```python
# Setup model architecture with trainable top layers
inputs = layers.Input(shape=(512,512,3), name="input_layer")
x = base_model_2(inputs, training=False) #Put the base model in inference mode, and so weights ehich remains frozen, remains frozen
x = layers.Conv2D(64, (3, 3), activation = 'relu')(x)
x = layers.Flatten()(x)
outputs = layers.Dense(len(class_names), activation="softmax", name="output_layer")(x)
model = tf.keras.Model(inputs, outputs)
```

```python
model.compile(loss = 'categorical_crossentropy', 
              optimizer = tf.keras.optimizers.Adam(learning_rate=0.0001), 
              metrics = ["accuracy"])
```

```python
# Fit
history = model.fit(data,
                    epochs=10,
                    steps_per_epoch=len(data))
```

```python
model.evaluate(data)
```
